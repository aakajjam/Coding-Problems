```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3.14, 3.14, 200) # This gives an array of points between -3.14 and 3.14 which is 200 numbers long
x2 = np.array([-3,-2,-1,0,1,2,3]) # This is seven numbers
```

```python
plt.plot(x, np.sin(x), c = 'red', linestyle = "--", linewidth = 5) # The c is the varible used to indicate color in this case red
# Takes 2 parameters. The first is an array 200 numbers long, the second is our y value (x,y plane)
# This is the sin(x) which is also the 200 numbers long between -3.14 and 3.14
plt.plot(x, np.cos(x), c = 'g', linestyle = ":", linewidth = 2) # You can abbreviate colors
# linestyle just indicates the shape of the line
# linewidth will make the lines look wider

plt.scatter(x2, x2**3/27, c = 'purple', marker = '3') 
# This gives a scatter plot of our x2 values. This is a graph so it takes an x and y value
# The marker is like the linestyle
plt.xlabel('X axis', fontsize = 20)
plt.ylabel('Height', fontsize = 6)
plt.title("Chart Title", c = 'red', fontweight = 'bold') 
# Fontweight can make it bold and fontsize changes the size of text
```
