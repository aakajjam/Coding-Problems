```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movieFile = pd.read_csv("movies.csv")
```

```python
movieFile.Rating = movieFile.Rating.replace('TV-PG', 'PG')
```

```python
movieFile.Rating
```

```python
plt.bar(movieFile.Rating, movieFile.GrossSales)
plt.xlabel('Ratings', c = 'red')
plt.ylabel('Gross Sales', c = 'green')
plt.title('Ratings and Gross Sales Comparison', c = 'purple')
```

```python
plt.scatter(movieFile.Budget, movieFile.GrossSales)
plt.xlabel('Budget')
plt.ylabel('Gross Sales')
plt.title('Budget and Gross Sales Plot')
```
