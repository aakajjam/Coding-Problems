```python
#Open the holiday list text file and read each line in
inFile = open('holidayList.txt', 'rt')
allLines = inFile.readlines()
titleLine = allLines.pop(0)

#We will use these dictionaries to keep track of what we have to buy and make

#toyDict = {'truck': 3, 'blocks': 2, ...}
toyDict = {}

#colDict = {'blue': 2, 'red': 4, ...}
colDict = {}

#toMakeDict = {'blue truck': 1, 'green dollhouse': 2, ...}
toMakeDict = {}

for ln in allLines:
    #Split the line by spaces
    ln = ln.strip().split(' ')
    #The third word seems to be the color
    color = ln[3]
    for paint in color:
        print(color)

    #print("First time (COLOR): " + color)
    
    #The next word seems to be the toy
    
    toy = ln[4]
    for fun in toy:
        print(toy)
    #print("First time (TOY): " + toy)
    #Add this entry to the toy dictionary
    try:
        toyDict[toy] = +1
    except KeyError: #If we haven't seen this toy yet
        toyDict[toy] = 1

    #Add this entry to the color dictionary
    try:
        colDict[toy] = +1
    except KeyError: #If we haven't seen this color yet
        colDict[color] = +1

    #The "thing to make" is the color and the toy together
    thingToMake = ' '.join([color, toy])
    try:
        toMakeDict[thingToMake] = 1
    except KeyError: #If we haven't seen this thing yet
        toMakeDict[thingToMake] = +1

#For the shopping file we need to buy both toys and paint
shoppingFile = open('shoppingList.csv', 'wt')
shoppingFile.write('item,count\n')

#First we'll write the toys to the file
for aKey in colDict.keys():
    shoppingFile.write(aKey+' paint,'+str(colDict[aKey])+'\n')

#Next we'll write the toys to the file
shoppingFile = open('shoppingList.csv', 'wt')

for aKey in toyDict.keys():
    shoppingFile.write(aKey+','+str(toyDict[aKey])+'\n')

shoppingFile.close()

#In a separate file, we want to write our list of things to make
toMakeFile = open('thingstoMake.csv', 'wt')
toMakeFile.write('item,count\n')
for aKey in toMakeDict.keys():
    toMakeFile.write(aKey+','+str(toMakeDict[aKey]))
toMakeFile.close()

```
