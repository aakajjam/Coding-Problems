```python
import numpy as np
import pandas as pd
```

```python
myData = pd.read_csv('favoriteMovies2022_raw.csv')
```

```python
myData
```

```python
myData.dtypes # dtype means data type
```

```python
for aCol in myData.columns:
    print(str(sum(myData[aCol].isna())) + " missing values for column " + aCol)
    # This isna() searches for missing values
```

```python
myData['Budget'] = myData['Budget'].fillna(np.nanmedian(myData['Budget'])) 
# Instead of filling missing budget with zero it fills the missing budgets with median budget value
# The .fillna method returns a new column, so we assign it to the same column/original data sheet
```

```python
# When you take a median value of a list or array with missing values, the answer is a missing value or NaN
# Having nanmedian will give the median of a list and ignore the NaN values
```

```python
print(myData.Rating.value_counts()) 
myData.Rating = myData.Rating.replace('R ', 'R')
myData.Rating = myData.Rating.replace('PG ', 'PG')
print(myData.Rating.value_counts())
# The value_counts gives the various counts of the Ratings
# The reason there are two "R" values is because one of them has a trailing space. That space is counted
```

```python
print(myData['TomatometerScore'].describe()) # Can also say myData.TomatometerScore.describe(), the mean the same thing
fixedData = myData.loc[myData.TomatometerScore > 1.0, 'TomatometerScore'] / 100 # The red text means the TomatometerScore column
# Looks at all the rows in which the TomatometerScore is greater than 1
myData.loc[myData.TomatometerScore > 1.0, 'TomatometerScore'] = fixedData 
# Take every row where the TomatometerScore was wrong and replace it with our fixed data
# The .loc means location
print(myData.TomatometerScore.describe())
```
